```python
# https://towardsdatascience.com/question-answering-with-pretrained-transformers-using-pytorch-c3e7a44b4012
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import datetime
from numpy.core.defchararray import add
```

```python
def create_df(i,q,a,s):
    data={'q_index': i,'question': q, 'answers' : a, 'sentence':s } 
    df=pd.DataFrame(data)
    return df
```

```python
def populate_answers(index, data):
    q=[]
    a=[]
    q.append('What is the call number?')
    a.append(data['call_number'])
    q.append('What call number?')
    a.append(data['call_number'])
    q.append('Which unit is assigned?')
    a.append(data['unit_id'])
    q.append('What unit is it?')
    a.append(data['unit_id'])
    q.append('What is the incident number?')
    a.append(data['incident_number'])
    q.append('What incident?')
    a.append(data['incident_number'])
    q.append('What type of incident is it?')
    a.append([data['call_type'],data['call_type_group']])
    q.append('What call type?')
    a.append([data['call_type'],data['call_type_group']])
    q.append('What is the reason for the call?')
    a.append([data['call_type'],data['call_type_group']])
    q.append('When was the call recieved?')
    a.append(data['call_date'])
    q.append('When was it reported?')
    a.append(data['call_date'])
    q.append('When was the incident submitted?')
    a.append(data['entry_dttm'])
    q.append('When did the dispatcher submit the task?')
    a.append(data['entry_dttm'])
    q.append('When did they respond?')
    a.append(data['response_dttm'])
    q.append('When did the unit acknowledge they were on route?')
    a.append(data['response_dttm'])
    q.append('When did they arrive on scene?')
    a.append(data['on_scene_dttm'])
    q.append('When did they arrive?')
    a.append(data['on_scene_dttm'])
    q.append('When did they get there?')
    a.append(data['on_scene_dttm'])
    q.append('What was the disposition?')
    a.append(data['call_final_disposition'])
    q.append('Why did it end?')
    a.append(data['call_final_disposition'])
    q.append('When did the unit leave?')
    a.append(data['available_dttm'])
    q.append('When did they leave?')
    a.append(data['available_dttm'])
    q.append('When did it end?')
    a.append(data['available_dttm'])
    q.append('What address?')
    a.append(data['address'])
    q.append('Where is it?')
    a.append([data['address'],data['neighborhoods_analysis_boundaries']])
    q.append('What is the location?')
    a.append([data['address'],data['neighborhoods_analysis_boundaries']])
    q.append('What is the zipcode?')
    a.append(data['zipcode_of_incident'])
    q.append('What zip?')
    a.append(data['zipcode_of_incident'])
    q.append('What battalion?')
    a.append(data['battalion'])
    q.append('What district?')
    a.append([data['station_area'],data['fire_prevention_district'],data['supervisor_district']])
    q.append('Which area?')
    a.append([data['station_area'],data['fire_prevention_district'],data['supervisor_district']])
    q.append('What jurisiction?')
    a.append([data['station_area'],data['fire_prevention_district'],data['supervisor_district']])
    q.append('What priroty?')
    a.append(data['priority'])
    q.append('Does this unit includes ALS (Advance Life Support) resources?')
    a.append(data['als_unit']) 
    q.append('Is there a paramedic in this unit?')
    a.append(data['als_unit']) 
    q.append('How many alarms?')
    a.append(data['number_of_alarms'])  
    q.append('Number of alarms?')
    a.append(data['number_of_alarms'])
    q.append('What type of unit?')
    a.append(data['unit_type']) 
    q.append('What sequence number?')
    a.append(data['unit_sequence_in_call_dispatch']) 
    q.append('What time did they arrive at the hospital?')
    a.append(data['hospital_dttm']) 
    i=[]
    s=[]
    q_num=0
    for item in q:
        i.append(str(index)+'_'+str(q_num))
        q_num+=1
        s.append(data['sentence'])
        
    df=create_df(i,q,a,s)
    
    return df
    
    
```

```python
def transform_data(data):
    data['call_date']= pd.to_datetime(data['call_date'])
    data['watch_date']= pd.to_datetime(data['watch_date'])
    data['received_dttm']= pd.to_datetime(data['received_dttm'])
    data['entry_dttm']= pd.to_datetime(data['entry_dttm'])
    data['dispatch_dttm']= pd.to_datetime(data['dispatch_dttm'])
    data['response_dttm']= pd.to_datetime(data['response_dttm'])
    data['on_scene_dttm']= pd.to_datetime(data['on_scene_dttm'])
    data['transport_dttm']= pd.to_datetime(data['transport_dttm'])
    data['hospital_dttm']= pd.to_datetime(data['hospital_dttm'])
    col_heads = data.columns.to_numpy().astype(str)
    data=data.astype(str)
    text_df = pd.DataFrame(add(add(col_heads, ' '), data), data.index, data.columns)
    data['sentence'] = text_df.apply('|'.join, axis=1)
    
    return data
    
```

```python
# set 10 day lookback window
periodend = datetime.datetime.now().isoformat()
periodstart = (datetime.datetime.now()- datetime.timedelta(days=10)).isoformat()
#Get count of incidents to determine limit to retrieve
num=pd.read_json("https://data.sfgov.org/resource/RowID.json?$select=COUNT(incident_number)&$where=call_date%20between%20%27"
                 +str(periodstart)+"%27%20AND%20%27"+str(periodend)+"%27")
maxNumber=num.iloc[0,0]
print('The number of incidents in the set is',maxNumber)
# Use record to write query call to API to get all incidents needed/bypass API default of 1000
query_str=("https://data.sfgov.org/resource/RowID.json?$where=call_date%20between%20%27"
           +str(periodstart)+"%27%20AND%20%27"
           +str(periodend)+"%27&$limit="+str(maxNumber))
cfs_data=pd.read_json(query_str)
cfs_data=cfs_data.dropna()
cfs_data.head()
```

```python
cfs_sample = cfs_data.sample(1000)
```

```python
model_name = "distilbert-base-cased-distilled-squad"
short_model_name = "distilbert"
qa = pipeline(task= "question-answering", model=model_name, tokenizer=model_name )
```

```python
cols = ['q_index','question','answers','sentence']
eval_set =pd.DataFrame(columns=cols)

cfs=transform_data(cfs_sample)
for index, row in cfs.iterrows():
    result = populate_answers(index,row)
    combine=[result, eval_set]
    eval_set = pd.concat(combine, ignore_index=True)

eval_set['ml_answer']=None
eval_set['ml_score']= None

for index, row in eval_set.iterrows():
    result = qa(question=row['question'], context=row['sentence'])
    eval_set.at[index,'ml_answer'] = result['answer']
    eval_set.at[index,'ml_score'] = result['score']
    
eval_set['score'] = eval_set.apply(lambda x: x.ml_answer in x.answers, axis=1)

eval_set.head()
```

```python
df_name = short_model_name+'_baselineEval.csv'
eval_set.to_csv(df_name, index=False)
```
